```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

#Fall

```python
#走路时滑倒导致向后摔倒:F02
df1= pd.read_csv('raw_data\F04_SE06_R04.txt',
                 header = None,usecols=[0,1,2], names = ["Ax", "Ay", "Az"])

#对df1 降采样，每两个数据取一个
df1=df1.loc[::2] #选取df1的奇数行
#重置索引
df1.index=list(range(df1.shape[0]))
Sensor1_Resolution = 13
Sensor1_Range = 16
g_S1 = (2*Sensor1_Range/2**Sensor1_Resolution)

df1['Ax']=g_S1*df1['Ax']
df1['Ay']=g_S1*df1['Ay']
df1['Az']=g_S1*df1['Az']

A1=[]

for i in range(0,df1.shape[0]):
    A1.append(np.sqrt(df1.iloc[i]['Ax']**2+df1.iloc[i]['Ay']**2+
                      df1.iloc[i]['Az']**2))
plt.figure(figsize=(15,9))
ax1 = plt.subplot(2, 2, 1)
plt.plot(A1,c='k')
plt.title('SMV')

ax1 = plt.subplot(2, 2, 2)
plt.plot(df1['Ax'],c='r')
plt.title('Ax')

ax1 = plt.subplot(2, 2, 3)
plt.plot(df1['Ay'],c='b')
plt.title('Ay')

ax1 = plt.subplot(2, 2, 4)
plt.plot(df1['Az'],c='g')
plt.title('Az')

plt.savefig("111.png")
plt.show();
```
